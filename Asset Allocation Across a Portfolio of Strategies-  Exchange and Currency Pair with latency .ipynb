```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Sample data including LSTM strategy
data = {
    'Strategy': ['Trend-Following', 'Mean-Reversion', 'Arbitrage', 'Trend-Following', 'Arbitrage', 'LSTM'],
    'Exchange': ['NYSE', 'NASDAQ', 'NYSE', 'NASDAQ', 'NASDAQ', 'NASDAQ'],
    'Currency Pair': ['EUR/USD', 'GBP/JPY', 'USD/JPY', 'EUR/USD', 'GBP/JPY', 'EUR/USD'],
    'Allocation': [30, 20, 25, 15, 10, 5]  # Allocation percentages with LSTM included
}

df_allocation = pd.DataFrame(data)

# Group by Exchange and Currency Pair and sum the Allocations
allocation_summary = df_allocation.groupby(['Exchange', 'Currency Pair']).sum().reset_index()
print("Allocation Summary:\n", allocation_summary)

# Create a pivot table
pivot_table = df_allocation.pivot_table(index='Exchange', columns='Currency Pair', values='Allocation', aggfunc='sum', fill_value=0)
print("Pivot Table:\n", pivot_table)

# Calculate total allocation for each group
total_allocation = df_allocation.groupby(['Exchange', 'Currency Pair'])['Allocation'].sum().reset_index(name='Total Allocation')

# Merge with the original data
df_merged = pd.merge(df_allocation, total_allocation, on=['Exchange', 'Currency Pair'])

# Calculate proportion
df_merged['Proportion'] = df_merged['Allocation'] / df_merged['Total Allocation']
print("Proportions:\n", df_merged)

# Plot allocation by Exchange and Currency Pair
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='g')
plt.title('Asset Allocation by Exchange and Currency Pair')
plt.show()

# Load data
def load_data(ticker, start_date, end_date):
    data = yf.download(ticker, start=start_date, end=end_date)
    return data

# Preprocess data
def preprocess_data(data):
    data = data[['Close', 'Volume']].dropna()
    data['Date'] = data.index
    data.set_index('Date', inplace=True)
    return data

# Plot predictions
def plot_predictions(y_test, y_pred, model_name):
    plt.figure(figsize=(12, 6))
    if isinstance(y_test, np.ndarray):
        y_test = pd.Series(y_test, index=y_pred.index if hasattr(y_pred, 'index') else range(len(y_test)))
    if isinstance(y_pred, np.ndarray):
        y_pred = pd.Series(y_pred, index=y_test.index if hasattr(y_test, 'index') else range(len(y_pred)))
    plt.plot(y_test.index, y_test, label=f'Actual Price ({model_name})')
    plt.plot(y_pred.index, y_pred, label=f'Predicted Price ({model_name})', linestyle='--')
    plt.title(f'{model_name} Price Prediction - Actual vs Predicted')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Train and evaluate LSTM model
def train_lstm_model(data):
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(data[['Close']])
    X = data_scaled
    y = data_scaled
    X = X.reshape((X.shape[0], 1, X.shape[1]))  # Reshape for LSTM input
    y = y.reshape((y.shape[0], 1))  # Reshape for LSTM output
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
    model = Sequential()
    model.add(LSTM(100, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True))
    model.add(LSTM(200))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(X_train, y_train, epochs=40, batch_size=1, verbose=1)
    y_pred = model.predict(X_test)
    y_test_inv = scaler.inverse_transform(y_test)
    y_pred_inv = scaler.inverse_transform(y_pred)
    mae = mean_absolute_error(y_test_inv, y_pred_inv)
    rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))
    print(f"LSTM Price Prediction - MAE: {mae}")
    print(f"LSTM Price Prediction - RMSE: {rmse}")
    y_test_inv = pd.Series(y_test_inv.flatten(), index=np.arange(len(y_test_inv)))
    y_pred_inv = pd.Series(y_pred_inv.flatten(), index=np.arange(len(y_pred_inv)))
    plot_predictions(y_test_inv, y_pred_inv, 'LSTM')

# Mean Reversion Strategy
def mean_reversion_strategy(data, window=20, z_score_threshold=1.5):
    data['Rolling Mean'] = data['Close'].rolling(window=window).mean()
    data['Rolling Std'] = data['Close'].rolling(window=window).std()
    data['Z-Score'] = (data['Close'] - data['Rolling Mean']) / data['Rolling Std']
    
    data['Signal'] = 0
    data.loc[data['Z-Score'] > z_score_threshold, 'Signal'] = -1  # Sell signal
    data.loc[data['Z-Score'] < -z_score_threshold, 'Signal'] = 1  # Buy signal
    
    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data['Close'], label='Close Price')
    plt.plot(data.index, data['Rolling Mean'], label='Rolling Mean', linestyle='--')
    plt.title('Mean Reversion Strategy')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    print(data[['Close', 'Rolling Mean', 'Z-Score', 'Signal']].tail())

# Trend Following Strategy
def trend_following_strategy(data, short_window=40, long_window=100):
    data['SMA40'] = data['Close'].rolling(window=short_window).mean()
    data['SMA100'] = data['Close'].rolling(window=long_window).mean()
    
    data['Signal'] = 0
    data.loc[data['SMA40'] > data['SMA100'], 'Signal'] = 1  # Buy signal
    data.loc[data['SMA40'] < data['SMA100'], 'Signal'] = -1  # Sell signal
    
    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data['Close'], label='Close Price')
    plt.plot(data.index, data['SMA40'], label=f'SMA {short_window}', linestyle='--')
    plt.plot(data.index, data['SMA100'], label=f'SMA {long_window}', linestyle='--')
    plt.title('Trend Following Strategy')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    print(data[['Close', 'SMA40', 'SMA100', 'Signal']].tail())

# Analyze anomalies: Too profitable or lossy trades, high latencies
def analyze_anomalies(data, profit_threshold=0.02, loss_threshold=-0.02):
    # Calculate daily returns
    data['Return'] = data['Close'].pct_change()
    
    # Identify anomalous returns
    data['Profit/Loss'] = data['Return'].apply(lambda x: 'Profitable' if x > profit_threshold else ('Lossy' if x < loss_threshold else 'Normal'))
    
    # Plot distribution of returns
    plt.figure(figsize=(12, 6))
    sns.histplot(data['Return'], bins=50, kde=True)
    plt.title('Distribution of Daily Returns')
    plt.xlabel('Return')
    plt.ylabel('Frequency')
    plt.show()
    
    # Print anomalies
    print("Anomalous Returns:")
    print(data[data['Profit/Loss'] != 'Normal'].tail())

    # Simulate latency data (for demonstration)
    np.random.seed(0)
    data['Latency'] = np.random.normal(loc=0.1, scale=0.02, size=len(data))  # Simulated latency data in seconds
    
    # Analyze latency
    latency_threshold = data['Latency'].mean() + 2 * data['Latency'].std()  # Define high latency as more than 2 std deviations from the mean
    data['Latency_Anomaly'] = data['Latency'] > latency_threshold
    
    # Plot latency distribution
    plt.figure(figsize=(12, 6))
    sns.histplot(data['Latency'], bins=50, kde=True)
    plt.axvline(latency_threshold, color='r', linestyle='--', label='High Latency Threshold')
    plt.title('Distribution of Latencies')
    plt.xlabel('Latency (seconds)')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
    
    # Print latency anomalies
    print("Latency Anomalies:")
    print(data[data['Latency_Anomaly']].tail())

# Main workflow
def main():
    ticker = 'AAPL'
    start_date = '2024-01-01'
    end_date = '2024-08-01'
    
    data = load_data(ticker, start_date, end_date)
    data = preprocess_data(data)
    
    train_lstm_model(data)
    
    # Add Mean Reversion and Trend Following Strategies
    mean_reversion_strategy(data)
    trend_following_strategy(data)
    
    # Analyze anomalies
    analyze_anomalies(data)

if __name__ == "__main__":
    main()
```
